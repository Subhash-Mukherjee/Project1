**Weather Patterns**

```
import io
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
fig,ax=plt.subplots()
```

```
seattle_weather=pd.read_csv(io.BytesIO(uploaded['seattle_weather.csv']))
```

```
austin_weather=pd.read_csv(io.BytesIO(uploaded['austin_weather.csv']))
```

```
fig,ax=plt.subplots()
ax.plot(seattle_weather["MONTH"],seattle_weather["MLY-PRCP-NORMAL"],color="b",marker="o")
ax.plot(austin_weather["MONTH"],austin_weather["MLY-PRCP-NORMAL"],color="r",marker="v")
plt.show()
```

```
abc,pqr=plt.subplots()
pqr.plot(seattle_weather["MONTH"],seattle_weather["MLY-PRCP-NORMAL"],color="b",marker="o",linestyle="--")
pqr.plot(austin_weather["MONTH"],austin_weather["MLY-PRCP-NORMAL"],color="r",marker="v",linestyle="--")
pqr.set_xlabel("Time (months)")
pqr.set_ylabel("weather patterns in austin and seattle")
plt.show()
```
